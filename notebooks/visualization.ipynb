```python
import pandas as pd
import matplotlib.pyplot as plt

# Load results
df = pd.read_csv("../models/model_comparison_results.csv", keep_default_na=False)

print(df)
```

```python
plt.figure(figsize=(8,5))

plt.bar(df["Model"], df["Test RMSE"])

plt.title("Model Comparison (Test RMSE)")
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.savefig("../reports/figures/model_comparison_rmse.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(df["Model"], df["R2"])

plt.title("Model Comparison (R² Score)")
plt.xlabel("Model")
plt.ylabel("R²")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(df["Model"], df["MAE"])

plt.title("Model Comparison (MAE)")
plt.xlabel("Model")
plt.ylabel("MAE")

plt.show()
```
