```python
plt.figure(figsize=(8,5))
sns.histplot(data["SalePrice"], kde=True)
plt.title("Sale Price Distribution")
plt.show()
```

```python
data["SalePrice"].skew()
```

```python
corr = data.corr(numeric_only=True)
plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
corr["SalePrice"].sort_values(ascending=False)
```

```python
top_corr = corr["SalePrice"].abs().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,6))
sns.barplot(x=top_corr.values, y=top_corr.index)
plt.title("Top Correlated Features With SalePrice")
plt.show()
```

```python
features = ["GrLivArea", "TotalBsmtSF", "GarageArea", "LotArea"]
for feature in features:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=data[feature], y=data["SalePrice"])
    plt.title(f"{feature} vs SalePrice")
    plt.show()
```

```python
categorical_features = ["OverallQual", "Neighborhood", "KitchenQual", "GarageType"]
for feature in categorical_features:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=data[feature], y=data["SalePrice"])
    plt.title(f"{feature} vs SalePrice")
    plt.xticks(rotation=45)
    plt.show()
```

```python
numeric_cols = data.select_dtypes(include=np.number).columns
data[numeric_cols].hist(figsize=(20,15))
plt.show()
```
